# **Persiapan Awal**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur tema visualisasi
sns.set_theme(style='whitegrid', palette='Set2')

# Memuat dataset
df_penguins = sns.load_dataset('penguins').dropna()
```

**Grafik 1: Bar Chart (Jumlah Spesies Pinguin)**

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.countplot(data=df_penguins, x='species', ax=ax)
ax.set_title('Jumlah Pinguin per Spesies')
ax.set_xlabel('Spesies')
ax.set_ylabel('Jumlah')
plt.show()
```

**Grafik 2: Histogram (Distribusi Panjang Sirip / Flipper Length)**

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.histplot(data=df_penguins, x='flipper_length_mm', kde=True, ax=ax)
ax.set_title('Distribusi Panjang Sirip Pinguin')
ax.set_xlabel('Panjang Sirip (mm)')
ax.set_ylabel('Frekuensi')
plt.show()
```

**Grafik 3: Boxplot (Distribusi Berat Badan per Spesies)**

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(data=df_penguins, x='species', y='body_mass_g', ax=ax)
ax.set_title('Distribusi Berat Badan Pinguin per Spesies')
ax.set_xlabel('Spesies')
ax.set_ylabel('Berat Badan (gram)')
plt.show()
```

**Grafik 4: Scatter Plot (Panjang Paruh vs Panjang Sirip)**

```python
fig, ax = plt.subplots(figsize=(9, 6))
sns.scatterplot(data=df_penguins, x='bill_length_mm', y='flipper_length_mm', hue='species', ax=ax)
ax.set_title('Hubungan Panjang Paruh dan Panjang Sirip Pinguin')
ax.set_xlabel('Panjang Paruh / Bill Length (mm)')
ax.set_ylabel('Panjang Sirip / Flipper Length (mm)')
plt.show()
```
