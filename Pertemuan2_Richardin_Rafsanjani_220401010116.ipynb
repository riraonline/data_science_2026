**2. Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%)
per kolom?**

```python
missing = df.isnull().sum()
percent = (missing / len(df)) * 100

pd.DataFrame({
    'Missing': missing,
    'Percent (%)': percent
})
```

**Jawaban:**

Kolom yang bermasalah utama:

*   Age - 177 - 19,9%
*   Cabin - 687 - 77,1%
*   Embarked - 2 - 0,22%





**3. Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy
atau Pandas.**

```python
survival_rate = df['Survived'].mean() * 100
survival_rate
```

**Jawaban:**

Mayoritas penumpang sebanyak 38.38%, sementara sisa nya tidak selamat sebanyak ±61%

**4. Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-
rata usianya?**

```python
female_p1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

count = female_p1.shape[0]
avg_age = female_p1['Age'].mean()

count, avg_age
```

**Jawaban:**

*   Jumlah: 94 orang
*   Rata-rata usia: ±34,6 tahun



**5. Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas
penumpang. Kelas mana yang paling tinggi?**

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean()
survival_by_class
```

**Jawaban:**

Kelas 1 cukup di pastikan selamat dan hal ini bisa menunjukkan bahwa semakin kaya seseorang maka semakin tinggi peluang selamat nya
